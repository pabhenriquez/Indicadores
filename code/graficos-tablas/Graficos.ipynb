# COVID-19

```python
df_covid = pd.read_excel('../../Resultados_covid.xlsx')
df_covid['nombre'] = '@'+ df_covid['nombre']
df_covid
```

```python
df_covid['engage_post']=(df_covid['engage_post']-df_covid['engage_post'].min())/(df_covid['engage_post'].max()-df_covid['engage_post'].min())
```

```python
fig = px.bar(df_covid, x='nombre', y='engage_post', color='Categoria', width=1200, height=500)
fig.update_layout(yaxis_title = 'Engagement', xaxis_title = '' ,
    font = dict(
        size = 20,
        color = '#000000'
    ),
    # autosize = True,
    legend=dict(
        x=0.01,
        y=0.98,
        title_text='',
        traceorder='normal',
        font=dict(
            size=20,),
    ),
    margin = dict(
        l = 10,
        b = 10,
        r = 10,
        t = 10
    ), 
    # legend_title_text = 'Type of Organization'
)
fig.show()
#fig.write_image('user-impact.pdf', engine='kaleido')
#fig.write_image('user-impact.png', engine='kaleido')
# fig.write_image('user-impact.html', engine='kaleido')
fig.write_image('Engagement_Covid_total.pdf', engine='kaleido')
```

```python
fig = px.bar(df_covid.groupby(["Categoria"]).mean(), x=df_covid.groupby(["Categoria"]).mean().index, y='engage_post', color=df_covid.groupby(["Categoria"]).mean().index, width=1200, height=500)
fig.update_layout(yaxis_title = 'Engagement', xaxis_title = '' ,
    font = dict(
        size = 20,
        color = '#000000'
    ),
    # autosize = True,
    margin = dict(
        l = 10,
        b = 10,
        r = 10,
        t = 10
    ), 
    showlegend = False
)
fig.show()
#fig.write_image('user-impact.pdf', engine='kaleido')
#fig.write_image('user-impact.png', engine='kaleido')
# fig.write_image('user-impact.html', engine='kaleido')
fig.write_image('Engagement_Covid_Category.pdf', engine='kaleido')
```
