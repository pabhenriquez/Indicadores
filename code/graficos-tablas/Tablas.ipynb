```python
import pandas as pd
import glob
```

```python
INCENDIO_2017_PATH = '../../data/incendios_2017/con_sentimientos/'
INCENDIO_2023_PATH = '../../data/incendios_2023/con_sentimientos/'
COVID_PATH = '../../data/covid/con_sentimientos/'
ESTALLIDO_PATH = '../../data/con_sentimientos_estallido/'

PATHS = [INCENDIO_2017_PATH, INCENDIO_2023_PATH, COVID_PATH, ESTALLIDO_PATH]
FECHAS = ['2017-01-01T00:00:00Z','2023-01-01T00:00:00Z', '2020-03-01T00:00:00Z', '2019-10-18T00:00:00Z'] 
```

```python
for i in range(len(PATHS)):
    path_prensa = PATHS[i] + 'prensa_escrita/' 
    path_television = PATHS[i] + 'television/' 
    path_influencers = PATHS[i] + 'influencers/' 

    print(i)
    for file in glob.glob(path_prensa+'/*.csv'):
        user_df = pd.read_csv(file)
   
        print('prensa, ' + user_df['username'].unique()[0]+ ', ' +str(len(user_df[user_df['created_at'] > FECHAS[i]])))
    
    for file in glob.glob(path_television+'/*.csv'):
        user_df = pd.read_csv(file)
   
        print('television, ' + user_df['username'].unique()[0]+ ', ' +str(len(user_df[user_df['created_at'] > FECHAS[i]])))
    
    for file in glob.glob(path_influencers+'/*.csv'):
        user_df = pd.read_csv(file)
   
        print('influencers, ' + user_df['username'].unique()[0]+ ', ' +str(len(user_df[user_df['created_at'] > FECHAS[i]])))
```
